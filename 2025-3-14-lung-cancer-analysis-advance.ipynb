# <p style="padding:10px;background-color:orange;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Correlation Heatmap</p>

Let's convert the target feature **"LUNG_CANCER"** from **Categorical** to **Numerical** data type by using **Label Encoder**. And converting the **"GENDER"** column from **Categorical** to **Numerical** data type by using **"One Hot Encoder"** for avoiding unexpected gender bias.

```python
from sklearn.preprocessing import LabelEncoder
LabelEncoder = LabelEncoder()

data["GENDER"] = data["GENDER"].replace({"M" : "Male" , "F" : "Female"})
data["LUNG_CANCER"] = LabelEncoder.fit_transform(data["LUNG_CANCER"])

data = pd.get_dummies(data, columns= ["GENDER"])
data.rename(columns={"GENDER_Male" : "MALE", "GENDER_Female" : "FEMALE", "YELLOW_FINGERS" : "YELLOW FINGERS", "PEER_PRESSURE" : "PEER PRESSURE", "LUNG_CANCER" : "LUNG CANCER", "FATIGUE " : "FATIGUE", "ALLERGY " : "ALLERGY"}, inplace=True)
data = data[["AGE","MALE","FEMALE","ALCOHOL CONSUMING","CHEST PAIN","SHORTNESS OF BREATH","COUGHING","PEER PRESSURE","CHRONIC DISEASE","SWALLOWING DIFFICULTY","YELLOW FINGERS","ANXIETY","FATIGUE","ALLERGY","WHEEZING","LUNG CANCER"]]
data.head().style.set_properties(**{"background-color": "#2a9d8f","color":"white","border": "1.5px  solid black"})
```

```python
plt.subplots(figsize =(16, 12))

p=sns.heatmap(data.corr(), cmap = palette, square=True, cbar_kws=dict(shrink =.99), 
            annot=True, vmin=-1, vmax=1, linewidths=0.1,linecolor='white',annot_kws=dict(fontsize =12))
p.axes.set_title("Pearson Correlation Of Features\n", fontsize=25)
plt.xticks(rotation=90)
plt.show()
```

<a id="13"></a>
# <p style="padding:10px;background-color:orange;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Preprocessing For Classification</p>

```python
x = data.drop("LUNG CANCER", axis = 1)
y = data["LUNG CANCER"]

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
x = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

print(f"Shape of training data : {x_train.shape}, {y_train.shape}")
print(f"Shape of testing data : {x_test.shape}, {y_test.shape}")
```

<a id="14"></a>
# <p style="padding:10px;background-color:orange;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Logistic Regression Model</p>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
lr_conf = confusion_matrix(y_test, lr_pred)
lr_report = classification_report(y_test, lr_pred)
lr_acc = round(accuracy_score(y_test, lr_pred)*100, ndigits = 2)
print(f"Confusion Matrix : \n\n{lr_conf}")
print(f"\nClassification Report : \n\n{lr_report}")
print(f"\nThe Accuracy of Logistic Regression is {lr_acc} %")
```

<a id="15"></a>
# <p style="padding:10px;background-color:orange;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Gaussian Naive Bayes Model</p>

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(x_train, y_train)
gnb_pred = gnb.predict(x_test)
gnb_conf = confusion_matrix(y_test, gnb_pred)
gnb_report = classification_report(y_test, gnb_pred)
gnb_acc = round(accuracy_score(y_test, gnb_pred)*100, ndigits = 2)
print(f"Confusion Matrix : \n\n{gnb_conf}")
print(f"\nClassification Report : \n\n{gnb_report}")
print(f"\nThe Accuracy of Gaussian Naive Bayes is {gnb_acc} %")
```

<a id="16"></a>
# <p style="padding:10px;background-color:orange;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Bernoulli Naive Bayes Model</p>

```python
from sklearn.naive_bayes import BernoulliNB

bnb = BernoulliNB()
bnb.fit(x_train, y_train)
bnb_pred = bnb.predict(x_test)
bnb_conf = confusion_matrix(y_test, bnb_pred)
bnb_report = classification_report(y_test, bnb_pred)
bnb_acc = round(accuracy_score(y_test, bnb_pred)*100, ndigits = 2)
print(f"Confusion Matrix : \n\n{bnb_conf}")
print(f"\nClassification Report : \n\n{bnb_report}")
print(f"\nThe Accuracy of Bernoulli Naive Bayes is {bnb_acc} %")
```

# <p style="padding:10px;background-color:orange;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Support Vector Machine Model</p>

```python
from sklearn.metrics import roc_curve, auc
from sklearn.utils import resample
import numpy as np

plt.figure(figsize=(12, 10))

# 
models = [
    ('Logistic Regression', lr, lr_pred),
    ('Gaussian Naive Bayes', gnb, gnb_pred),
    ('Bernoulli Naive Bayes', bnb, bnb_pred),
    ('Support Vector Machine', svm, svm_pred),
    ('Random Forest', rfg, rfg_pred),
    ('K-Nearest Neighbors', knn, knn_pred),
    ('XGBoost', xgb, xgb_pred)
]

# bootstrap
n_bootstraps = 1000
confidence_level = 0.95
colors = plt.cm.tab10(np.linspace(0, 1, len(models)))

# y_testy_scorenumpy
y_test_array = np.array(y_test)

# ROC
for idx, (name, model, pred) in enumerate(models):
    # 
    try:
        y_score = model.predict_proba(x_test)[:, 1]
    except:
        # SVM
        try:
            y_score = model.decision_function(x_test)
        except:
            # 
            y_score = pred
    
    # y_scorenumpy
    y_score_array = np.array(y_score)
    
    # ROC
    fpr, tpr, _ = roc_curve(y_test_array, y_score_array)
    roc_auc = auc(fpr, tpr)
    
    # ROC
    plt.plot(fpr, tpr, lw=2, color=colors[idx], label=f'{name} (AUC = {roc_auc:.2f})')
    
    # Bootstrap
    tprs = []
    aucs = []
    
    # FPRTPR
    base_fpr = np.linspace(0, 1, 100)
    
    for i in range(n_bootstraps):
        # 
        indices = resample(range(len(y_test_array)), replace=True, n_samples=len(y_test_array))
        if len(np.unique(y_test_array[indices])) < 2:
            # 
            continue
        
        # bootstrapROC
        fpr_boot, tpr_boot, _ = roc_curve(y_test_array[indices], y_score_array[indices])
        aucs.append(auc(fpr_boot, tpr_boot))
        
        # FPRTPR
        tpr_interp = np.interp(base_fpr, fpr_boot, tpr_boot)
        tpr_interp[0] = 0.0  # 
        tprs.append(tpr_interp)
    
    # TPR
    tprs = np.array(tprs)
    mean_tprs = tprs.mean(axis=0)
    std_tprs = tprs.std(axis=0)
    
    tprs_upper = np.minimum(mean_tprs + 1.96 * std_tprs, 1)
    tprs_lower = np.maximum(mean_tprs - 1.96 * std_tprs, 0)
    
    # AUC
    auc_mean = np.mean(aucs)
    auc_std = np.std(aucs)
    auc_lower = max(0, auc_mean - 1.96 * auc_std)
    auc_upper = min(1, auc_mean + 1.96 * auc_std)
    
    # 
    plt.fill_between(base_fpr, tprs_lower, tprs_upper, color=colors[idx], alpha=0.2,
                    label=f'95% CI [{auc_lower:.2f}-{auc_upper:.2f}]')

# 
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('Comparison of ROC Curves for Various Models (with 95% Confidence Intervals)', fontsize=18)
plt.legend(loc="lower right", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import auc

# Set figure size
plt.figure(figsize=(12, 10))

# Function to calculate cumulative gains
def cumulative_gain_curve(y_true, y_score):
    # Order by decreasing probability
    desc_score_indices = np.argsort(y_score)[::-1]
    y_true_sorted = y_true[desc_score_indices]
    
    # Calculate cumulative number of positives
    cum_positives = np.cumsum(y_true_sorted)
    
    # Calculate percentage of total positives
    positives_ratio = cum_positives / float(np.sum(y_true))
    
    # Calculate percentage of samples
    percentage_of_samples = np.arange(1, len(y_true) + 1) / float(len(y_true))
    
    # Add the origin point (0,0)
    percentage_of_samples = np.insert(percentage_of_samples, 0, 0)
    positives_ratio = np.insert(positives_ratio, 0, 0)
    
    return percentage_of_samples, positives_ratio

# Calculate and plot for each model
y_test_array = np.array(y_test)

for idx, (name, model, pred) in enumerate(models):
    # Get prediction scores
    try:
        y_score = model.predict_proba(x_test)[:, 1]
    except:
        try:
            y_score = model.decision_function(x_test)
        except:
            y_score = pred
    
    # Convert to numpy array
    y_score_array = np.array(y_score)
    
    # Calculate cumulative gains
    percentages, gains = cumulative_gain_curve(y_test_array, y_score_array)
    
    # Calculate AUC for the gains curve
    gains_auc = auc(percentages, gains)
    
    # Plot curve
    plt.plot(percentages, gains, lw=2, 
             label=f'{name} (AUC = {gains_auc:.2f})')

# Add perfect model line
percents, gains = cumulative_gain_curve(y_test_array, y_test_array)
plt.plot(percents, gains, 'k-.', lw=2, label='Perfect Model')

# Add random model line (diagonal)
plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Model')

# Set plot properties
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Percentage of Samples', fontsize=14)
plt.ylabel('Percentage of Positive Class', fontsize=14)
plt.title('Cumulative Gains Chart', fontsize=18)
plt.legend(loc="lower right", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(12, 10))

# Function to calculate lift
def lift_curve(y_true, y_score):
    # Get the indices that would sort the predictions in descending order
    desc_score_indices = np.argsort(y_score)[::-1]
    y_true_sorted = y_true[desc_score_indices]
    
    # Number of positive instances
    n_pos = np.sum(y_true)
    
    # Total number of instances
    n_instances = len(y_true)
    
    # Random selection baseline (constant)
    baseline = n_pos / n_instances
    
    # Calculate lift for different percentiles
    percentiles = np.arange(0, n_instances + 1) / n_instances
    lifts = np.zeros(n_instances + 1)
    lifts[0] = 1.0  # Start with lift = 1
    
    # Calculate lift for each percentile
    for i in range(1, n_instances + 1):
        n_pos_in_percentile = np.sum(y_true_sorted[:i])
        lifts[i] = (n_pos_in_percentile / i) / baseline if i > 0 else 1.0
    
    return percentiles, lifts

# Calculate and plot lift for each model
for idx, (name, model, pred) in enumerate(models):
    # Get prediction scores
    try:
        y_score = model.predict_proba(x_test)[:, 1]
    except:
        try:
            y_score = model.decision_function(x_test)
        except:
            y_score = pred
    
    # Convert to numpy array
    y_score_array = np.array(y_score)
    
    # Calculate lift curve
    percentiles, lifts = lift_curve(y_test_array, y_score_array)
    
    # Plot lift curve
    plt.plot(percentiles, lifts, lw=2, label=name)

# Add baseline (random model)
plt.axhline(y=1.0, color='k', linestyle='--', label='Random Model')

# Set plot properties
plt.xlim([0.0, 1.0])
plt.ylim(bottom=0.5)  # Adjust as needed based on your lift values
plt.xlabel('Percentage of Samples', fontsize=14)
plt.ylabel('Lift Value', fontsize=14)
plt.title('Lift Chart', fontsize=18)
plt.legend(loc="best", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```
